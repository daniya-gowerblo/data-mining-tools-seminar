```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], -1) / 255.0
X_test = X_test.reshape(X_test.shape[0], -1) / 255.0

# В keras датасет хранится уже в преобразованном числовом формате
X_train[0]
```

```python
encoder = OneHotEncoder(sparse_output=False)
y_train_onehot = encoder.fit_transform(y_train.reshape(-1, 1))
y_test_onehot = encoder.transform(y_test.reshape(-1, 1))
```

# Класс линейного слоя

```python
class Linear:
    def __init__(self, input_size, output_size, learning_rate=0.01):
        self.weights = np.random.randn(input_size, output_size) * 0.01
        self.bias = np.zeros((1, output_size))
        self.input = None
        self.learning_rate = learning_rate

    def forward(self, x):
        self.input = x
        return np.dot(x, self.weights) + self.bias

    def backward(self, grad_output):
        grad_input = np.dot(grad_output, self.weights.T)
        grad_weights = np.dot(self.input.T, grad_output)
        grad_bias = np.sum(grad_output, axis=0, keepdims=True)

        # Обновление параметров
        self.weights -= self.learning_rate * grad_weights
        self.bias -= self.learning_rate * grad_bias

        return grad_input
```

# Класс активаций функций

```python
class ReLU:
    def forward(self, x):
        self.input = x
        return np.maximum(0, x)

    def backward(self, grad_output):
        return grad_output * (self.input > 0)


class Softmax:
    def forward(self, x):
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        self.output = exp_x / np.sum(exp_x, axis=1, keepdims=True)
        return self.output

    def backward(self, grad_output):
        y = self.output
        sum_yg = np.sum(y * grad_output, axis=1, keepdims=True)
        return y * (grad_output - sum_yg)
```

# Класс для функций потерь MSE

```python
class MSELoss:
    def forward(self, y_pred, y_true):
        self.y_pred = y_pred
        self.y_true = y_true
        return np.mean((y_pred - y_true) ** 2)

    def backward(self):
        return 2 * (self.y_pred - self.y_true) / self.y_true.shape[0]
```

# Архитектура нейронной сети

```python
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size, learning_rate=0.01):
        self.linear1 = Linear(input_size, hidden_size, learning_rate)
        self.relu = ReLU()
        self.linear2 = Linear(hidden_size, output_size, learning_rate)
        self.softmax = Softmax()
        self.loss_fn = MSELoss()

    def forward(self, x):
        out = self.linear1.forward(x)
        out = self.relu.forward(out)
        out = self.linear2.forward(out)
        out = self.softmax.forward(out)
        return out

    def backward(self, y_true):
        grad_loss = self.loss_fn.backward()
        grad_softmax = self.softmax.backward(grad_loss)
        grad_linear2 = self.linear2.backward(grad_softmax)
        grad_relu = self.relu.backward(grad_linear2)
        self.linear1.backward(grad_relu)
```

# Обучение сети

```python
input_size = 28*28
hidden_size = 128
output_size = 10

learning_rate = 0.01
num_epochs = 15
batch_size = 64
```

```python
model = NeuralNetwork(input_size, hidden_size, output_size, learning_rate=learning_rate)

for epoch in range(num_epochs):
    for i in range(0, len(X_train), batch_size):
        X_batch = X_train[i:i + batch_size]
        y_batch = y_train_onehot[i:i + batch_size]

        y_pred = model.forward(X_batch)

        loss = model.loss_fn.forward(y_pred, y_batch)
        model.backward(y_batch)

    # if epoch / 10 == int(epoch / 10):
    print(f'Epoch {epoch + 0}/{num_epochs}, Loss: {loss}')
```

```python
predictions = model.forward(X_test)

predicted_classes = np.argmax(predictions, axis=1)
accuracy = np.mean(predicted_classes == y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
def plot_images(images, n_cols=10):
    n_images = len(images)
    n_rows = (n_images + n_cols - 1) // n_cols
    plt.figure(figsize=(n_cols * 2, n_rows * 2))
    for i in range(n_images):
        plt.subplot(n_rows, n_cols, i + 1)
        plt.imshow(images[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
    plt.show()

images_to_display = X_test[:30]
plot_images(images_to_display)
```
